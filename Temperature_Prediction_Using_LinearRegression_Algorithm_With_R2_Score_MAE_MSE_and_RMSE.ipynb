```python
#dropping_tables
df=df.drop(['Precip Type','Loud Cover'],axis=1)
```

```python
columns=df.select_dtypes(include=['object']).columns
#Label_Encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
for col in columns:
  df[col]=a.fit_transform(df[col])
df
```

```python
#heatmap
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.show()
```

```python
x=df.drop('Temperature (C)',axis=1)
y=df['Temperature (C)']
```

```python
#Simple_Linear_Regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
#accuracy will be predicted directly in Simple_Linear_Regression
reg.score(x,y)
```

```python
#y=mx+c
#coefficient(m)
reg.coef_
```

```python
#intercept(c)
reg.intercept_
```

```python
#training_and_testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred1=reg.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
#r2_score
r2_score(y_test,y_pred1)
```

```python
#mean_absolute_error
mean_absolute_error(y_test,y_pred1)
```

```python
#mean_squared_error
mean_squared_error(y_test,y_pred1)
```

```python
#root_mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred1))
```
